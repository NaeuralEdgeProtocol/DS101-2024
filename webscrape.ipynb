```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import requests
from bs4 import BeautifulSoup
```

```python
# create urls for all seasons of all leagues
base_url = 'https://understat.com/league'
leagues = ['La_liga', 'EPL', 'Bundesliga', 'Serie_A', 'Ligue_1', 'RFPL']
seasons = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']
```

```python
import json

string_with_json_obj = ''

# Find data for teams
for el in scripts:
    if 'teamsData' in el.text:
      string_with_json_obj = el.text.strip()
      
# print(string_with_json_obj)

# strip unnecessary symbols and get only JSON data
ind_start = string_with_json_obj.index("('")+2
ind_end = string_with_json_obj.index("')")
json_data = string_with_json_obj[ind_start:ind_end]

json_data = json_data.encode('utf8').decode('unicode_escape')
```

```python
from tqdm import tqdm  # Import tqdm

full_data = dict()

# Add tqdm for leagues
for league in tqdm(leagues, desc="Processing Leagues"):
    season_data = dict()
    
    # Add tqdm for seasons within each league
    for season in tqdm(seasons, desc=f"Processing {league} seasons", leave=False):
        url = base_url + '/' + league + '/' + season
        res = requests.get(url)
        soup = BeautifulSoup(res.content, "lxml")

        # Find data in <script> tags
        scripts = soup.find_all('script')
        string_with_json_obj = ''
        
        # Extract JSON data
        for el in scripts:
            if 'teamsData' in el.text:
                string_with_json_obj = el.text.strip()
        
        ind_start = string_with_json_obj.index("('") + 2
        ind_end = string_with_json_obj.index("')")
        json_data = string_with_json_obj[ind_start:ind_end]
        json_data = json_data.encode('utf8').decode('unicode_escape')
        
        # Convert JSON to dict
        data = json.loads(json_data)
        
        # Get teams and IDs
        teams = {id: data[id]['title'] for id in data.keys()}
        
        # Get column names from first team's data
        columns = list(data[next(iter(data.keys()))]['history'][0].keys())
        
        # Process team data
        dataframes = {}
        for id, team in teams.items():
            teams_data = [list(row.values()) for row in data[id]['history']]
            df = pd.DataFrame(teams_data, columns=columns)
            dataframes[team] = df
        
        # Add PPDA metrics
        for team, df in dataframes.items():
            df['ppda_coef'] = df['ppda'].apply(lambda x: x['att']/x['def'] if x['def'] != 0 else 0)
            df['ppda_att'] = df['ppda'].apply(lambda x: x['att'])
            df['ppda_def'] = df['ppda'].apply(lambda x: x['def'])
            df['oppda_coef'] = df['ppda_allowed'].apply(lambda x: x['att']/x['def'] if x['def'] != 0 else 0)
            df['oppda_att'] = df['ppda_allowed'].apply(lambda x: x['att'])
            df['oppda_def'] = df['ppda_allowed'].apply(lambda x: x['def'])
        
        # Combine all team data
        frames = []
        for team, df in dataframes.items():
            df['team'] = team
            frames.append(df)
        
        full_stat = pd.concat(frames).drop(['ppda', 'ppda_allowed'], axis=1)
        full_stat['xG_diff'] = full_stat['xG'] - full_stat['scored']
        full_stat['xGA_diff'] = full_stat['xGA'] - full_stat['missed']
        full_stat['xpts_diff'] = full_stat['xpts'] - full_stat['pts']
        full_stat.reset_index(drop=True, inplace=True)
        
        season_data[season] = full_stat

    # Combine seasonal data for the league
    df_season = pd.concat(season_data)
    full_data[league] = df_season

# Final combined data
data = pd.concat(full_data)
data.head()
```

```python
data.index = data.index.droplevel(2)
data.index = data.index.rename(names=['league','year'], level=[0,1])
data.head()
```
