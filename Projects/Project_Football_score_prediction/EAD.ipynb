# Load and visualize dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("merged_matches.csv")

df.head()
```

```python
df.info()
```

```python
print(df.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder_league = LabelEncoder()
df['league'] = encoder_league.fit_transform(df['league'])

encoder_h_a_home = LabelEncoder()
df['h_a_home'] = encoder_h_a_home.fit_transform(df['h_a_home'])

encoder_result_home = LabelEncoder()
df['result_home'] = encoder_result_home.fit_transform(df['result_home'])

encoder_h_a_away = LabelEncoder()
df['h_a_away'] = encoder_h_a_away.fit_transform(df['h_a_away'])

encoder_result_away = LabelEncoder()
df['result_away'] = encoder_result_away.fit_transform(df['result_away'])

encoder_team_home = LabelEncoder()
df['team_home'] = encoder_team_home.fit_transform(df['team_home'])

encoder_team_away = LabelEncoder()
df['team_away'] = encoder_team_away.fit_transform(df['team_away'])

df['date'] = pd.to_datetime(df['date'])
df['date'] = (df['date'] - pd.Timestamp("2014-08-23")).dt.days 

df['result'] = df['result_home']

team_name = "La_liga"
team_encoded = encoder_league.transform([team_name])[0]
print(f"Encoding pentru {team_name}: {team_encoded}")

df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df.corr()

fig, ax = plt.subplots(figsize=(100, 100))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, ax=ax)

plt.savefig("correlation_matrix.png", dpi=300, bbox_inches='tight')
```

# Create a new df based on original one

```python
# Drop the columns that have a low correlation with result and use just one if a column has a strong correlation with another one
```

```python
df = pd.read_csv("merged_matches.csv")
```

```python
df_new = df[['date', 'team_home', 'team_away']].copy()
df_new['xG_home'] = df['xG_home']
df_new['xG_away'] = df['xG_away']
df_new['npxGD'] = df['npxGD_home']
df_new['deep_home'] = df['deep_home']
df_new['deep_away'] = df['deep_away']
df_new['xpts_diff'] = df['xpts_home']
df_new['xpts_away'] = df['xpts_away']
df_new['xG_diff_home'] = df['xG_diff_home']
df_new['xGA_diff_home'] = df['xGA_diff_home']
df_new['xpts_diff_home'] = df['xpts_diff_home']
df_new['xG_diff_away'] = df['xG_diff_away']
df_new['xGA_diff_away'] = df['xGA_diff_away']
df_new['xpts_diff_away'] = df['xpts_diff_away']
df_new['scored_home'] = df['scored_home']
df_new['scored_away'] = df['scored_away']
df_new['result'] = df['result_home']
```

```python
# Eliminate the duplicates
```

```python
df_new = df_new.drop_duplicates()
```

```python
df_new.head()
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

result_counts = df_new['result'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=result_counts.index, y=result_counts.values, hue=result_counts.index, palette="coolwarm")

plt.title("DistribuÈ›ia clasei 'result'")
plt.xlabel("Rezultat")
plt.ylabel("NumÄƒr de apariÈ›ii")

plt.show
```

```python
df_new.to_csv("clean_data.csv", index=False)
```

```python
df_encoded = df_new.copy()

df_encoded['date'] = pd.to_datetime(df_encoded['date'])
df_encoded['date'] = (df_encoded['date'] - pd.Timestamp("2014-08-23")).dt.days 

encoder_team_home = LabelEncoder()
df_encoded['team_home'] = encoder_team_home.fit_transform(df_encoded['team_home'])

encoder_team_away = LabelEncoder()
df_encoded['team_away'] = encoder_team_away.fit_transform(df_encoded['team_away'])

result_mapping = {'w': 1, 'l': 2, 'd': 3}
df_encoded['result'] = df_encoded['result'].map(result_mapping)

team_home_mapping = pd.DataFrame({
    'team_name': encoder_team_home.classes_,
    'team_home_encoded': range(len(encoder_team_home.classes_))
})

team_away_mapping = pd.DataFrame({
    'team_name': encoder_team_away.classes_,
    'team_away_encoded': range(len(encoder_team_away.classes_))
})

team_home_mapping.to_csv("team_home_mapping.csv", index=False)
team_away_mapping.to_csv("team_away_mapping.csv", index=False)

df_encoded.head()
```

```python
df_encoded.to_csv("clean_data_encoded.csv", index=False)
```
